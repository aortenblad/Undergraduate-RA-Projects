```python
df = pd.read_csv('Amendment_Detail.csv', encoding = "ISO-8859-1")
df.tail()
```

```python
df['Ammendments'] = df['Ammendments'].astype(str)
df.rename(columns={'Ammendments':'Amendments'}, inplace=True)
df.head()
```

```python
df['regex'] = df['Amendments'].str.findall(r'(\$[0-9,.]+ ?[m,b]?)')
df.head()
```

```python
def convert_into_words(list_of_char):
    return_list = []
    for i in list_of_char:
        if len(return_list) == 0:
            return_list.append(i)
        elif i == ' ':
            return_list.append('')
        else:
            return_list[-1] = return_list[-1] + i

    return return_list

convert_into_words(df['Amendments'][0])
```

```python
df['Separated'] = df['Amendments'].apply(convert_into_words)
df.head()
```

```python
Increase = ['increase', 'available', 'provide', 'increases', 'provides', 'appropriates', 'restores', 'restore', 'insert', 'inserts']
Decrease = ['decrease', 'reduce', 'cut', 'halt', 'reduces', 'strike', 'strikes', 'save', 'eliminate', 'eliminating', 'deletes', 'delete', 'limit', 'limits', 'eliminates', 'end']
Neutral = ['earmarks', 'earmark', 'transfer', 'transfers', 'redirect', 'redirects']

def change(list):
    list_of_changes = []
    for x in list:
        if x in Increase:
            list_of_changes.append('Increase')
        elif x in Decrease:
            list_of_changes.append('Decrease')
        elif x in Neutral:
            list_of_changes.append('Neutral')
    return list_of_changes
```

```python
change(df['Separated'][23])
```

```python
df['Change'] = df['Separated'].apply(change)
df.head()
```

```python
# here we convert all the dollar amounts into a standard format; that is, into their integer representations

for i in df['regex']:
    for j in range(len(i)): 
        if i[j][-1] == 'm':
            i[j] = (int(i[j].replace("$", "").replace(",", "").replace(".", "").replace(" m", "")) * 1000000)
        elif i[j][-1] == 'b':
            i[j] = (int(i[j].replace("$", "").replace(",", "").replace(".", "").replace(" b", "")) * 1000000000)
        else:
            i[j] = (int(i[j].replace("$", "").replace(",", "").replace(".", "")))
```
