```python
import pandas as pd
import numpy as np
import re
```

```python
def get_years2(substring): #deprecated version, see below for latest function
    if type(substring) is not str:
        return None # substring not string type
        
    years = re.search(r"(\d{4})-(\d{4})", substring)
    if years:
        print(years.group(0))
        return list(range(int(years.group(1)), int(years.group(2)) + 1))
    
    years = re.search(r"(\d{4})-(\d{2})", substring)
    if years:
        print(years.group(0))
        return list(range(int(years.group(1)), int(years.group(1)[0:2] + years.group(2)) + 1))
    
    years = re.search(r"\d{4}", substring)
    if years:
        print(years.group(0))
        return int(years.group(0))
    
    return None
```

```python
current_year = 2018

def get_years(substring): 
    if type(substring) is not str:
        return None # substring not string type
    substring = " " + substring + " "
    
    all_years = re.findall(r"\d{4}-\d{4}|\d{4}-\D|\d{4}-\d{2}\D|[^-]\d{4}[^-]", substring)
    if len(all_years) != 1:
        return None

    years = re.search(r"(\d{4})-(\d{4})", all_years[0])
    if years:
        print(years.group(0))
        return list(range(int(years.group(1)), int(years.group(2)) + 1))
    
    years = re.search(r"(\d{4})-\D", all_years[0])
    if years:
        print(years.group(0))
        return list(range(int(years.group(1)), current_year + 1))
    
    years = re.search(r"(\d{4})-(\d{2})\D", all_years[0])
    if years:
        print(years.group(0))
        return list(range(int(years.group(1)), int(years.group(1)[0:2] + years.group(2)) + 1))
    
    years = re.search(r"[^-](\d{4})[^-]", all_years[0])
    if years:
        print(years.group(0))
        return int(years.group(1))
    
    return None
```

```python
file_name = "cd_bios, mostly no duplicates_locations"

df = pd.read_csv(file_name + ".csv", index_col=0)

print("Applying function to dataframe...")
for i in range(1, 101):
    df["years" + str(i)] = df["substring" + str(i)].apply(get_years)

print("Finished: ")
display(df)
```

```python
bio_dta = pd.DataFrame(columns=["name", "year", "fips"])
```

```python
for row in range(len(df)):
    for i in range(1, 101):
        if type(df["years" + str(i)][row]) is list:
            if not np.isnan(df["fips" + str(i)][row]):
                for year in df["years" + str(i)][row]:
                    bio_dta = bio_dta.append(pd.DataFrame({"name": [df["name"][row]], "year": [year],
                                         "fips": [df["fips" + str(i)][row]]}))
        elif df["years" + str(i)][row] and not np.isnan(df["years" + str(i)][row]):
            if not np.isnan(df["fips" + str(i)][row]):
                bio_dta = bio_dta.append(pd.DataFrame({"name": [df["name"][row]], "year": [df["years" + str(i)][row]],
                                     "fips": [df["fips" + str(i)][row]]}))

bio_dta["year"] = bio_dta["year"].astype("int")
bio_dta["fips"] = bio_dta["fips"].astype("int")
bio_dta = bio_dta.groupby(["name", "year"])["fips"].apply(set).to_frame()
bio_dta = bio_dta.sort_values(["name", "year"])
display(bio_dta)
```
